```python
from gurobipy import *
 
m = Model("qcp")

 
a = m.addVar(name="I12")
b = m.addVar(name="I13")
c = m.addVar(name="I23")
d = m.addVar(name="I24")
e = m.addVar(name="I34")
 
obj = 1*a**2+ 4*b**2+ 6*c**2+ 12*d**2+ 3*e**2
m.setObjective(obj, GRB.MINIMIZE)
 
m.addConstr(a+ b == 710, "c0")
m.addConstr(a == c+d, "c1")
m.addConstr(b+ c== e, "c2")
m.addConstr(c+ e== 710, "c3")
 
m.addConstr(a>=0, "c4")
m.addConstr(b>=0, "c5")
m.addConstr(c>=0, "c6")
m.addConstr(d>=0, "c7")
m.addConstr(e>=0, "c8")

m.write("Q2.lp")

m.optimize()

for v in m.getVars():
  print('%s %.2f' % (v.varName, v.x))  

print('Obj: %.2f' % obj.getValue())

 
 

 


 

```
