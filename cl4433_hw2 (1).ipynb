```python
import requests
from time import sleep
from random import random
from datetime import datetime

def get_businesses(query, api_key, min_rating=0.0, max_results=50):
    base_url = "https://maps.googleapis.com/maps/api/place/textsearch/json"
    distance_url = "https://maps.googleapis.com/maps/api/distancematrix/json"
    origin_place_id = "ChIJyQ3Tlj72wokRUCflR_kzeVc"

    results = []
    next_page_token = None
    total_results = 0

    while True:
        params = {
            "query": query,
            "key": api_key,
            "minprice": 0,
            "maxprice": 4,
            "open_now": str(datetime.now().time())[:5],   
            "minrating": min_rating,
            "pagetoken": next_page_token
        }

        response = requests.get(base_url, params=params)
        data = response.json()

        if data['status'] != 'OK':
            break

        for place in data['results']:
            if total_results >= max_results:
                return results
 
            details_params = {
                "origins": f"place_id:{origin_place_id}",
                "destinations": f"place_id:{place['place_id']}",
                "key": api_key
            }

            distance_response = requests.get(distance_url, params=details_params)
            distance_data = distance_response.json()

            if distance_data['status'] == 'OK':
                total_results += 1
                result = (
                    place['name'],
                    place['formatted_address'],
                    {'open_now_1': place.get('opening_hours', {}).get('open_now', False)},
                    place.get('price_level', None),
                    place.get('rating', None),
                    distance_data['rows'][0]['elements'][0]['distance']['text'],
                    distance_data['rows'][0]['elements'][0]['duration']['text']
                )
                
                results.append(result)
        #print(results)
        next_page_token = data.get('next_page_token')

        if not next_page_token or total_results >= max_results:
            break
     
        sleep(2 + random())
    filtered_results = [x for x in results if x[4] >= min_rating]
    
    def convert_to_meters(distance):
        if 'km' in distance:
            return float(distance.split()[0]) * 1000
        elif 'm' in distance:
            return float(distance.split()[0])
        else:
            return float('inf')   
 
    def custom_sort(x):
        rating = -x[4]
        distance = convert_to_meters(x[5])
        return (rating, distance)
 
    filtered_results = sorted(filtered_results, key=custom_sort)
        
            
            
    return filtered_results[:max_results]
 
api_key = 'AIzaSyCcfT2kolWXBrIa8T8__uJpVrcLoUgQf8Q'
sorted_results = get_businesses("restaurants near Columbia University", api_key, min_rating=0.0, max_results=50)
print(sorted_results)
```
